```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
```

```python
wine_df = pd.read_csv('WineQT.csv')
wine_df
```

```python
wine_df.head()
```

```python
wine_df.info()
```

```python
# Descriptive Statistics of the Dataset
wine_df.describe()
```

```python
# Check for missing values
missing_values = wine_df.isnull().sum()
```

```python
# Check for duplicate entries
duplicates = wine_df.duplicated().sum()
print("Number of duplicate rows:", duplicates)
```

```python
# Histograms
wine_df.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# Box plots
plt.figure(figsize=(12, 6))
sns.boxplot(x='quality', y='fixed acidity', data=wine_df)
plt.title('Fixed Acidity vs Quality')
plt.show()
```

```python
# Calculate the average quality for each level of alcohol
alcohol_quality = wine_df.groupby('alcohol')['quality'].mean().reset_index()
```

```python
# Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(alcohol_quality['alcohol'], alcohol_quality['quality'], marker='o', linestyle='-', color='b')
plt.title('Average Quality vs Alcohol Content')
plt.xlabel('Average Quality')
plt.ylabel('Alcohol')
plt.grid(True)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(wine_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='alcohol', y='quality', data=wine_df, hue='quality', palette='viridis')
plt.title('Alcohol vs Quality')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='quality', y='volatile acidity', data=wine_df)
plt.title('Mean Volatile Acidity by Quality')
plt.show()
```
